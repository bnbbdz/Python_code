### About dataset


This dataset is about past loans. The **Loan_train.csv** data set includes details of 346 customers whose loan are already paid off or defaulted. It includes following fields:

| Field          | Description                                                                           |
| -------------- | ------------------------------------------------------------------------------------- |
| Loan_status    | Whether a loan is paid off on in collection                                           |
| Principal      | Basic principal loan amount at the                                                    |
| Terms          | Origination terms which can be weekly (7 days), biweekly, and monthly payoff schedule |
| Effective_date | When the loan got originated and took effects                                         |
| Due_date       | Since it’s one-time payoff schedule, each loan has one single due date                |
| Age            | Age of applicant                                                                      |
| Education      | Education of applicant                                                                |
| Gender         | The gender of applicant                                                               |

### Convert to date time object


```python
df['due_date'] = pd.to_datetime(df['due_date'])
df['effective_date'] = pd.to_datetime(df['effective_date'])
df.head()
```

# Pre-processing:  Feature selection/extraction


### Let's look at the day of the week people get the loan


```python
df['dayofweek'] = df['effective_date'].dt.dayofweek
bins = np.linspace(df.dayofweek.min(), df.dayofweek.max(), 10)
g = sns.FacetGrid(df, col="Gender", hue="loan_status", palette="Set1", col_wrap=2)
g.map(plt.hist, 'dayofweek', bins=bins, ec="k")
g.axes[-1].legend()
plt.show()
```

```python
df_test = df
df_test['dow'] = df_test['due_date'].dt.dayofweek
bins_1 = np.linspace(df_test.dow.min(),df_test.dow.max(),10)
g_test = sns.FacetGrid(df_test,col='Gender',hue = 'loan_status',palette = 'Set1',col_wrap = 2)
g_test.map(plt.hist,'dow',bins=bins,ec='k')
g_test.axes[0].legend()
plt.show()
```

We see that people who get the loan at the end of the week don't pay it off, so let's use Feature binarization to set a threshold value less than day 4


```python
df['weekend'] = df['dayofweek'].apply(lambda x: 1 if (x>3)  else 0)
df.head()
```

## Convert Categorical features to numerical values


Let's look at gender:


```python
df.groupby(['Gender'])['loan_status'].value_counts(normalize=True)
```

86 % of female pay there loans while only 73 % of males pay there loan


Let's convert male to 0 and female to 1:


```python
df['Gender'].replace(to_replace=['male','female'], value=[0,1],inplace=True)
df.head()
```

## One Hot Encoding

#### How about education?


```python
df.groupby(['education'])['loan_status'].value_counts(normalize=True)
```

#### Features before One Hot Encoding


```python
df[['Principal','terms','age','Gender','education']].head()
```

#### Use one hot encoding technique to conver categorical varables to binary variables and append them to the feature Data Frame


```python
Feature = df[['Principal','terms','age','Gender','weekend']]
Feature = pd.concat([Feature,pd.get_dummies(df['education'])], axis=1)
Feature.drop(['Master or Above'], axis = 1,inplace=True)
Feature.head()
```

### Feature Selection


Let's define feature sets, X:


```python
X = Feature
X[0:5]
```

What are our lables?


```python
df['loan_status'].replace(to_replace = ['PAIDOFF','COLLECTION'],value = [0,1], inplace = True)
y = df['loan_status'].values
y[0:5]
```

## Normalize Data


Data Standardization give data zero mean and unit variance (technically should be done after train test split)


```python
X= preprocessing.StandardScaler().fit(X).transform(X)
X[0:5]
```

# K Nearest Neighbor(KNN)

Notice: You should find the best k to build the model with the best accuracy.\
**warning:** You should not use the **loan_test.csv** for finding the best k, however, you can split your train_loan.csv into train and test to find the best **k**.


```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
#Split train set and test set: 70% tran, 30% test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 4)
```

```python
#Find the best K for the model: from 1 to 15
ks = 15
k_mean_acc = np.zeros(ks-1)
k_std_acc = np.zeros(ks-1)

#loop through K value to find K
for k in range(1,ks):
    neigh = KNeighborsClassifier(n_neighbors = k).fit(X_train,y_train)
    y_hat = neigh.predict(X_test)
    k_mean_acc[k-1] = metrics.accuracy_score(y_test,y_hat)
    k_std_acc [k-1] = np.std(y_hat == y_test)/np.sqrt(y_hat.shape[0])

#plot K to see the best K easily    
plt.plot(range(1,ks),k_mean_acc,'r')
plt.fill_between(range(1,ks),k_mean_acc - 1*k_std_acc, k_mean_acc + 1*k_std_acc, alpha = 0.1)
plt.fill_between(range(1,ks),k_mean_acc - 3*k_std_acc, k_mean_acc + 3*k_std_acc, alpha = 0.1, color = 'y')
plt.xlabel('Number of K')
plt.ylabel('Accuracy')
plt.title('The accuracy and standard deviation through difference K')
plt.legend(('Accuracy','+/-1 Std','+/-3 Std'))
plt.tight_layout
plt.show()
```

```python
#As you can see on the chart above K = 2 is the best choice
k = 2
neigh = KNeighborsClassifier(n_neighbors = k).fit(X, y)
neigh
```

# Decision Tree


```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dc_tree = DecisionTreeClassifier(criterion = 'entropy',max_depth = 4)
tree = dc_tree.fit(X,y)
tree
```

# Support Vector Machine


```python
from sklearn import svm
```

```python
clf = svm.SVC(kernel = 'rbf')
```

```python
clf_svm = clf.fit(X,y)
clf_svm
```

# Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression(C=0.01, solver='liblinear')
```

```python
lrs = LR.fit(X,y)
lrs
```

# Model Evaluation using Test set


```python
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss
```

First, download and load the test set:


```python
!wget -O loan_test.csv https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/ML0101ENv3/labs/loan_test.csv
```

### Load Test set for evaluation


```python
test_df = pd.read_csv('loan_test.csv')
test_df.head()
```

**Preprocessing test data**

```python
test_df['due_date'] = pd.to_datetime(test_df['due_date'])
test_df['effective_date'] = pd.to_datetime(test_df['effective_date'])
test_df['dayofweek'] = test_df['effective_date'].dt.dayofweek
test_df['weekend'] = test_df['dayofweek'].apply(lambda x: 1 if (x>3)  else 0)
test_df['Gender'].replace(to_replace = ['male','female'], value = [0,1],inplace = True)
test_df['loan_status'].replace(to_replace = ['PAIDOFF','COLLECTION'],value = [0,1], inplace = True)

Feature_test = test_df[['Principal','terms','age','Gender','weekend']]
Feature_test = pd.concat([Feature_test,pd.get_dummies(test_df['education'])], axis=1)
Feature_test.drop(['Master or Above'], axis = 1, inplace = True)
X_valid = Feature_test
X_valid = preprocessing.StandardScaler().fit(X_valid).transform(X_valid)
y_valid = test_df['loan_status'].values
```

**Predict y_hat for each model**

```python
#KNN
y_hat_knn = neigh.predict(X_valid)

#Decision Tree
y_hat_tree = tree.predict(X_valid)

#Support Vector Machine
y_hat_svm = clf_svm.predict(X_valid)

#Logistic Regression
y_hat_lr = lrs.predict(X_valid)
y_hat_lr_prob = lrs.predict_proba(X_valid)
```

**Model evaluation**

KNN

```python
#Jaccard
Jaccard_knn = jaccard_score(y_valid, y_hat_knn, average='weighted')

#F1-score
F1_knn = metrics.f1_score(y_valid, y_hat_knn, average='weighted')

print('Jaccard for KNN: ',Jaccard_knn ,'\nF1-score for KNN: ',F1_knn)
```

Decision Tree

```python
#Jaccard
Jaccard_tree = jaccard_score(y_valid, y_hat_tree, average='weighted')

#F1-score
F1_tree = metrics.f1_score(y_valid, y_hat_tree, average='weighted')

print('Jaccard for Decision Tree: ',Jaccard_tree ,'\nF1-score for Decision Tree: ',F1_tree)
```

Support Vector Machine

```python
#Jaccard
Jaccard_svm = jaccard_score(y_valid, y_hat_svm, average='weighted')

#F1-score
F1_svm = metrics.f1_score(y_valid, y_hat_svm, average='weighted')

print('Jaccard for Support Vector Machine: ',Jaccard_svm ,'\nF1-score for Support Vector Machine: ',F1_svm)
```

Logistic Regression

```python
#Jaccard
Jaccard_lr = jaccard_score(y_valid, y_hat_lr, average='weighted')

#F1-score
F1_lr = metrics.f1_score(y_valid, y_hat_lr, average='weighted')

#LogLoss
Logloss_lr = metrics.log_loss(y_valid, y_hat_lr_prob)

print('Jaccard for Logistic Regression: ',Jaccard_lr ,'\nF1-score for Logistic Regression: ',F1_lr,'\nLog loss for Logistic Regression: ',Logloss_lr)
```

Summary

```python
Summary = pd.DataFrame(columns =[['Algorithm','Jaccard','F1-score','LogLoss']])
Summary_knn = pd.Series(['KNN',Jaccard_knn,F1_knn,np.nan], index = Summary.columns)
Summary_tree = pd.Series(['Decision Tree',Jaccard_tree,F1_tree,np.nan], index = Summary.columns)
Summary_svm = pd.Series(['SVM',Jaccard_svm,F1_svm,np.nan], index = Summary.columns)
Summary_lr = pd.Series(['LogisticRegression',Jaccard_lr,F1_lr,Logloss_lr], index = Summary.columns)

Summary = Summary.append(Summary_knn,ignore_index= True)
Summary = Summary.append(Summary_tree,ignore_index= True)
Summary = Summary.append(Summary_svm,ignore_index= True)
Summary = Summary.append(Summary_lr,ignore_index= True)
Summary
```
