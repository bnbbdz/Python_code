```python
# url = "https://karofionline.com/san-pham/"
urls =[]
for i in range(1,8):
    urls.append(f"https://karofionline.com/san-pham/page/{i}/")
```

```python
products = []
prices = []
links =[]
```

```python
for url in urls:
    data  = requests.get(url).text 
    soup = BeautifulSoup(data,"html5lib")
    meta = soup.find_all("meta", itemprop="price")
    product_link = soup.find_all("h3", itemprop="name")

    for i in meta:
        prices.append(i.get('content'))

    for i in product_link:
        link = i.find('a',href=True)
        links.append(link.get('href'))
        products.append(link.text)
```

```python
df = pd.DataFrame()
df['product'] = products
df['price'] = prices
df['link'] = links
df
```
