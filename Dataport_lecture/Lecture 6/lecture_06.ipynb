# Task 2

```python
df1 = pd.read_excel('lecture_06.xlsx', sheet_name = '2. Student_grade')
```

```python
df2 = pd.read_excel('lecture_06.xlsx', sheet_name = '2. Student_attendance')
```

```python
df3 = pd.read_excel('lecture_06.xlsx', sheet_name = '2. Student_attendance_status')
```

```python
formatted_df1 = pd.melt(df1,["Student"], var_name="Lecture A", value_name = 'Grade')
```

```python
formatted_df2 = pd.melt(df2,["Student"], var_name="Lecture B", value_name = 'Status')
```

```python
result = formatted_df1.merge(formatted_df2, how = 'left'
                             , left_on = ['Student', 'Lecture A']
                            , right_on = ['Student', 'Lecture B'])
```

```python
result
```

```python
result = result.merge(df3, how = 'left')
```

```python
result#.drop(columns = 'Status')
```

```python
columns = ['Student','Lecture','Status_Name','Grade']
```

```python
# result = result.reindex(columns=columns)
result
```

```python
import numpy as np
```

```python
result['is_attended'] = (result['Status'] == 1) | (result['Status'] == 2)
```

```python
group_result = result.groupby(['Student']).agg({'Grade':np.mean, 'is_attended':np.sum})
```
