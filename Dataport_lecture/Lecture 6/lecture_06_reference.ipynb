```python
messy = pd.DataFrame({'First' : ['John', 'Jane', 'Mary'], 
                      'Last' : ['Smith', 'Doe', 'Johnson'], 
                      'Treatment A' : [np.nan, 16, 3], 
                      'Treatment B' : [2, 11, 1]})
messy
```

```python
tidy = pd.melt(messy, 
               id_vars=['First','Last'], 
               var_name='treatment', 
               value_name='result')
tidy.dropna()
```

```python
# revert to origion by pivot table
pivot = tidy.pivot_table(index=['First', 'Last'], columns='treatment', 
                         values='result')
pivot
```

```python
pivot.columns.ravel( )
```

```python
pivot.columns = pivot.columns.ravel( )
pivot = pivot.reset_index()
pivot
```

# Reshaping Data 

### Pivot

```python
from collections import OrderedDict
table = OrderedDict((
    ("Item", ['Item0', 'Item0', 'Item1', 'Item1']),
    ('CType',['Gold', 'Bronze', 'Gold', 'Silver']),
    ('USD',  ['1$', '2$', '3$', '4$']),
    ('EU',   ['1€', '2€', '3€', '4€'])
))
df = pd.DataFrame(table)
df
```

```python
pivot_df = df.pivot(index='Item', columns='CType') 
# print(pivot_df.columns)
pivot_df
```

```python
pivot_df.columns = pivot_df.columns.ravel( )
pivot_df = pivot_df.reset_index()
pivot_df
```

# Combine Data Sets

```python
pd.merge?
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                       'key2': ['K0', 'K0', 'K0', 'K0'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})

print(left)
print(right)
result = pd.merge(left, right, on=['key1', 'key2'])
result
```

```python
pd.merge(left, right, how='left', left_on=['key1', 'key2'], right_on = ['key1', 'key2'])
```

```python
pd.merge(left, right, how='right', on=['key1', 'key2'])
```

```python
pd.merge(left, right, how='outer', on=['key1', 'key2'])
```

```python
print(left)
print(right)
left.join(right, lsuffix='_trai', rsuffix='_phai') #lsuffix='_left', rsuffix='_right'
```
