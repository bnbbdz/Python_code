```python
import pandas as pd
from statistics import mean
```

```python
criminal = pd.read_excel(r'C:\Users\manhcuong.nguyen\Dropbox\Python\Datapot\Lecture 6\lecture_06.xlsx',sheet_name = '1. Criminal_cases')
attendance = pd.read_excel(r'C:\Users\manhcuong.nguyen\Dropbox\Python\Datapot\Lecture 6\lecture_06.xlsx',sheet_name = '2. Student_attendance')
grade = pd.read_excel(r'C:\Users\manhcuong.nguyen\Dropbox\Python\Datapot\Lecture 6\lecture_06.xlsx',sheet_name = '2. Student_grade')
status = pd.read_excel(r'C:\Users\manhcuong.nguyen\Dropbox\Python\Datapot\Lecture 6\lecture_06.xlsx',sheet_name = '2. Student_attendance_status')
```

```python
criminal
```

```python
criminal_pivot = criminal.pivot_table(index =['country','year'], columns = 'type',values = 'count')
#reset index, inlace để thay đổi dataframe
criminal_pivot.reset_index(drop = False , inplace = True)
criminal_pivot
```

```python
attendance_melt = pd.melt(attendance, id_vars = ['Student'], var_name = 'Lecture A', value_name = 'Status')
attendance_melt.dropna()
```

```python
grade_melt = pd.melt(grade, id_vars = ['Student'], var_name = 'Lecture B', value_name = 'Grade')
grade_melt.dropna()
```

```python
df = pd.merge(attendance_melt, grade_melt, how = 'left' , left_on = ['Student','Lecture A'], right_on = ['Student','Lecture B'])
df
```

```python
result = df.merge(status , how = 'outer')
# result_new = result[['Student','Lecture A','Grade']]
```

```python
# result_new
result['is attendance'] = (result['Status'] == 1 ) | (result['Status'] == 2 )
result
```

```python
new = result.groupby(['Student']).agg({'Grade' : mean,'is attendance': sum})
```
