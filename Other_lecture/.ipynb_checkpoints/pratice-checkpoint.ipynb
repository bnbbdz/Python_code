```python
fish[fish['Country Name']=='VIETNAM']
```

```python
fish[
    (fish['Country Name']=='VIETNAM') &
    (fish['Feature']=='IMP Quantity') 
].sort_values(by=['Year', 'Month number']).tail(5)
```

```python
Tình huống 2: Lọc ra thông tin khối lượng nhập khẩu ("IMP Quantity") của Việt Nam ("VIETNAM"). 
    Sau đó dùng 3 cách khác nhau để xem 5 tháng gần nhất.
    
```

```python
fish[
    (fish['Country Name']=='VIETNAM') & 
    (fish['Feature']=='IMP Quantity')
].sort_values(by=['Year','Month number']).tail(5)[['Product Name', 'Feature']] 
```
